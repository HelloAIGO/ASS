```python
import os 
import sys 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from matplotlib import font_manager
from datetime import datetime, timedelta
import csv
from sklearn.datasets import load_iris
```

```python
# iris 데이터셋 로드
iris = load_iris()

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
```

```python
iris_df
```

```python
# Set the style for Seaborn
sns.set(style="whitegrid")

# Create a pair plot to visualize relationships between features
sns.pairplot(iris_df, hue="target", palette="viridis")

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(iris_df['sepal length (cm)'][iris_df['target'] == 0], iris_df['sepal width (cm)'][iris_df['target'] == 0], label="setosa", c="b")
plt.scatter(iris_df['sepal length (cm)'][iris_df['target'] == 1], iris_df['sepal width (cm)'][iris_df['target'] == 1], label="versicolor", c="g")
plt.scatter(iris_df['sepal length (cm)'][iris_df['target'] == 2], iris_df['sepal width (cm)'][iris_df['target'] == 2], label="virginica", c="r")

# Add labels and a legend
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Iris Dataset - Sepal Length vs. Sepal Width")
plt.legend(title="Species", loc="best")
```
